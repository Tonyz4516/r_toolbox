```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("CRAIGSLIST.Listings_modified_V2.csv")
```

```
df["LISTING_ID"] = [str(a).replace(".0", "") for a in df["LISTING_ID"]]
d = {}
for id in df["LISTING_ID"]:
    if id not in d:
        d[id] = 1
    else:
        d[id] += 1
```

```
count_of_repeat = []
for k in d:
    count_of_repeat.append(d[k])

import plotly.express as px
fig = px.histogram(x=count_of_repeat)
fig.show()
```

```
label = []
listing_ids = list(df["LISTING_ID"])
for i in range(df.shape[0] - 1, -1, -1):
    # loop back to front
    key = listing_ids[i]
    if d[key] > 1:
        d[key] -= 1
        label.append("duplicated")
    else:
        d[key] -= 1
        label.append("original")

df["label"] = label[::-1]
```

```
df.to_csv("CRAIGSLIST.Listings_modified_V3.csv")
```
